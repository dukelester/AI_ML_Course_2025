```python
# Import Matplotlib

%matplotlib inline
import matplotlib.pyplot as plt

# Prepare the data

x = np.random.rand(9, 400)
y = x **2

# Set the plot

fig, ax = plt.subplots(figsize=(10, 10))

# plot the data

ax.plot(x, y)

# Customize the figure

ax.set(title='Random Plot', xlabel='Random X', ylabel='Square of X')
```

```python
# Load Data

admission_data = pd.read_csv('HDHI Admission data.csv')
admission_data.head(10)
```

```python
age_above_50.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.set_xlabel('AGE')
ax.set_ylabel('GENDER')

age_above_50.plot(x='AGE', y='GENDER',kind='scatter', ax=ax)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the data 

netflix = pd.read_csv('data/netflix_titles.csv')
netflix.head(10)
```

```python
netflix.tail(10)
```

```python
netflix.describe()
```

```python
netflix.info()
```

```python

columns = netflix.columns
print('Total columns \t', columns, len(columns))

total_rows = len(list(netflix.index))
print('total rows \t', total_rows)
```

```python
# Missing data and Null Rate
result = {}
for column in columns:
    null_rate = netflix[column].isna().sum() / len(netflix) * 100
    
    if null_rate > 0:
        result[column] = f'{round(null_rate, 2)} %'
print(f'The Null Rate:  {result}')
```

```python
# Handle the missing data
country_mode = netflix['country'].mode()[0]
country_mode

# Replace the null country with modest -> United States

netflix['country'] = netflix['country'].fillna(country_mode)
netflix['cast'].replace(np.nan, 'No Cast', inplace=True)
netflix['director'].replace(np.nan, 'No Director', inplace=True)
netflix['rating'].replace(np.nan, 'PG', inplace=True)

# drop other Nan

netflix.dropna(inplace=True)

# Drop Duplicates

netflix.drop_duplicates(inplace=True)
```

```python
netflix.isnull().sum()
```

```python
# Fix the Date

netflix['date_added'] = pd.to_datetime(netflix['date_added'])
netflix['Month'] = netflix['date_added'].dt.month
netflix['Month Name'] = netflix['date_added'].dt.month_name()
netflix['Year'] = netflix['date_added'].dt.year

netflix.head(10)
```

## 2. Whatâ€™s the most common genre on Netflix?

Genres are stored in the listed_in column of Netflix dataset. Each row can have multiple genres separated by commas (e.g. "International TV Shows, Romantic TV Shows, TV Dramas").

To find the most common genre, weâ€™ll split them and count frequencies.

```python
netflix.head(10)
```

```python
listed_in = netflix['listed_in'].str.split(',').explode().str.strip()
listed_in
```

```python

listed_in.value_counts()
```

```python
genre_counts = listed_in.value_counts()
print("Top 10 Genres:\n", genre_counts.head(10), '\n\n')
print(f'The most common genre in Netflix is {genre_counts.idxmax()} with {genre_counts.max()}')
print(f'The least common genre in Netflix is {genre_counts.idxmin()} with {genre_counts.min()}')
```

```python
plt.figure(figsize=(15, 6))
genre_counts.head(10).plot(kind='bar', color='purple')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Top 10 Most Common Genres on Netflix')
plt.show()
```

5. Who are the top 5 most frequent directors on Netflix?


```python
netflix.head(4)
```

```python
directors = netflix['director']
top_directors = directors.value_counts().head(10)
top_directors

print("Top 5 most frequent directors on Netflix:\n")
print(top_directors.head(6))
```

```python
import pandas as pd
```

```python
data = pd.read_csv('data/netflix_titles.csv')
data.head(4)

# Drop rows with missing directors
all_directors = data.dropna(subset=['director'])
all_directors

movie_directors = all_directors[all_directors['type'] == 'Movie']
tv_shows_directors = all_directors[all_directors['type'] == 'TV Show']

top_movie_directors = movie_directors['director'].value_counts()
top_tv_shows_directors = tv_shows_directors['director'].value_counts()


print("ðŸŽ¬ Top 5 Movie Directors:\n", top_movie_directors.head(5))
print("\nðŸ“º Top 5 TV Show Directors:\n", top_tv_shows_directors.head(5))

# Plot side-by-side

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

top_movie_directors.head(10).plot(kind='bar', ax=axes[0], color='orangered')
axes[0].set_title('Top 5 Movie Directors In Netflix')
axes[0].set_xlabel('Director')
axes[0].set_ylabel('Number Of Movies')
axes[0].tick_params(axis='x', rotation=45)

top_tv_shows_directors.head(10).plot(kind='bar', ax=axes[1], color='purple')
axes[1].set_title('Top 5 TV Show Directors In Netflix')
axes[1].set_xlabel('Director')
axes[1].set_ylabel('Number Of TV Shows')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

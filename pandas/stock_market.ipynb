## Problem Statement:
An investor wants insights into stock price movements to support decision-making.

#### Questions to Solve:

    - What is the average closing price of the stock?

    - Which day had the highest trading volume?

### 1. What is the average closing price of the stock?

```python
tesla_stocks.head(10)
```

```python
mean_close = tesla_stocks['Close'].mean()
print(f"Average Closing Price: {mean_close:.2f}")
```

### 2. Which day had the highest trading volume?

```python
tesla_stocks.head()
```

```python
sorted_by_volume = tesla_stocks.sort_values(by='Volume', ascending=False)
sorted_by_volume
```

```python
print(f'{sorted_by_volume.iloc[0].Date} was the day with highest volume of {sorted_by_volume.iloc[0].Volume}')
```

```python
max_date = tesla_stocks.loc[tesla_stocks['Volume'].idxmax()]
max_date.Date

print(f'{max_date.Date} was the day with highest volume of {max_date.Volume}')
```

### 3. Create a new column for daily returns %.

```python
tesla_stocks.head()
```

```python
difference = tesla_stocks['Close'] - tesla_stocks['Open']
percentage = (difference / tesla_stocks['Close']) * 100

tesla_stocks['Percentage'] = percentage
tesla_stocks.head(10)
```

```python
difference
```

```python
tesla_stocks['Daily Returns %'] = tesla_stocks['Close'].pct_change() * 100
tesla_stocks.head(10)
```

### 4. Calculate the 7-day and 30-day moving averages of closing price.

```python
tesla_stocks['Moving AGV 7 Day'] = tesla_stocks['Close'].rolling(window=7, min_periods=1).mean()
tesla_stocks['Moving AGV 30 Day'] = tesla_stocks['Close'].rolling(window=30, min_periods=1).mean()

tesla_stocks.head(10)
```

5. Compare the average return in 2019 vs 2020.

```python
tesla_stocks['Date'] = pd.to_datetime(tesla_stocks['Date'])
tesla_stocks.head(9)

tesla_stocks['Year'] = tesla_stocks['Date'].dt.year
tesla_stocks.head(9)

group_by_year = tesla_stocks.groupby('Year')['Daily Returns %'].mean()
group_by_year

```

```python
avg_2019 = group_by_year.get(2019, None)
avg_2020 = group_by_year.get(2020, None)
```

```python
avg_2019, avg_2020
```

```python
print("Average Daily Return in 2019:", avg_2019)
print("Average Daily Return in 2020:", avg_2020)
```

## 1. Which product category generates the most revenue?

### 5. What percentage of orders are cancelled vs completed?

```python
ecommerce_data[ecommerce_data['Revenue'] < 0]
```

```python
ecommerce_data[ecommerce_data['Revenue'] > 0]
```

```python
ecommerce_data['Orderstatus'] = ecommerce_data['InvoiceNo'].str.startswith('C')
ecommerce_data['Orderstatus'] 
```

```python
ecommerce_data['Orderstatus'] = ecommerce_data['Orderstatus'].map({True: 'Cancelled', False: 'Completed'})
ecommerce_data
```

```python
order_counts = ecommerce_data.groupby('Orderstatus')['InvoiceNo'].nunique()
order_counts
```

```python
percentage = order_counts / order_counts.sum() * 100
percentage
```

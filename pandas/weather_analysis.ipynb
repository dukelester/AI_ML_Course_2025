1. Find the hotest and coldest date

```python
sorted_data = weather_data.sort_values(by="Temperature (C)", ascending=False)
sorted_data.head(2)
```

```python
sorted_data.tail(2)
```

```python
hottest = list(sorted_data.head(1)['Formatted Date'])[0]
coldest = list(sorted_data.tail(1)['Formatted Date'])[0]
```

```python
print(f'The hottest date is {hottest} and coldest date is {coldest}')
```

## 2. What is the average monthly temperature?

```python
sorted_data.groupby(['Formatted Date'])["Temperature (C)"].mean()
```

```python
monthly_avg = weather_data.groupby('Formatted Date')["Temperature (C)"].mean()

monthly_avg
```
